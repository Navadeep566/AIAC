```python
# Generate test cases
test_cases = [
    (2, True),  # Smallest prime
    (3, True),
    (4, False), # Smallest non-prime
    (5, True),
    (10, False),
    (11, True),
    (17, True),
    (20, False),
    (23, True),
    (100, False),
    (97, True), # A larger prime
]

# Implement is_prime function
def is_prime(n):
    """Checks if a number is prime.

    Args:
        n: An integer greater than 1.

    Returns:
        True if n is prime, False otherwise.
    """
    if n <= 1:
        return False # Constraint: input should be greater than 1
    if n <= 3:
        return True # 2 and 3 are prime
    if n % 2 == 0 or n % 3 == 0:
        return False # Divisible by 2 or 3

    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# Get user input
while True:
    try:
        user_input = int(input("Enter an integer greater than 1: "))
        if user_input > 1:
            break
        else:
            print("Input must be greater than 1. Please try again.")
    except ValueError:
        print("Invalid input. Please enter an integer.")

# Test the function
print("Test Cases:")
for num, expected in test_cases:
    result = is_prime(num)
    print(f"Input: {num}, Result: {result}, Expected: {expected}")

print(f"\nUser Input: {user_input}")
user_result = is_prime(user_input)
print(f"Input: {user_input}, Result: {user_result}")
```

```python
# Define the conversion functions

def celsius_to_fahrenheit(celsius):
    """Converts Celsius to Fahrenheit."""
    if not isinstance(celsius, (int, float)):
        return None
    return (celsius * 9/5) + 32

def fahrenheit_to_celsius(fahrenheit):
    """Converts Fahrenheit to Celsius."""
    if not isinstance(fahrenheit, (int, float)):
        return None
    return (fahrenheit - 32) * 5/9

# Define test cases for both functions
celsius_to_fahrenheit_test_cases = [
    (0, 32.0),
    (100, 212.0),
    (25.5, 77.9),
    (-10, 14.0),
    ("abc", None),
    (None, None),
    ([1, 2], None)
]

fahrenheit_to_celsius_test_cases = [
    (32, 0.0),
    (212, 100.0),
    (77.9, 25.5),
    (14, -10.0),
    ("xyz", None),
    (None, None),
    ({"a": 1}, None)
]

# Run tests for celsius_to_fahrenheit
print("Testing celsius_to_fahrenheit:")
for input_c, expected_f in celsius_to_fahrenheit_test_cases:
    result_f = celsius_to_fahrenheit(input_c)
    print(f"Input: {input_c}, Result: {result_f}, Expected: {expected_f}")

# Run tests for fahrenheit_to_celsius
print("\nTesting fahrenheit_to_celsius:")
for input_f, expected_c in fahrenheit_to_celsius_test_cases:
    result_c = fahrenheit_to_celsius(input_f)
    print(f"Input: {input_f}, Result: {result_c}, Expected: {expected_c}")

# Get user input and perform conversion
while True:
    try:
        user_temp_str = input("\nEnter a temperature (e.g., 25C or 77F): ")
        user_temp = float(user_temp_str[:-1])
        user_unit = user_temp_str[-1].upper()
        if user_unit in ['C', 'F']:
            break
        else:
            print("Invalid unit. Please enter 'C' for Celsius or 'F' for Fahrenheit.")
    except (ValueError, IndexError):
        print("Invalid input format. Please enter a number followed by 'C' or 'F'.")

print(f"\nUser Input: {user_temp_str}")
if user_unit == 'C':
    converted_temp = celsius_to_fahrenheit(user_temp)
    if converted_temp is not None:
        print(f"{user_temp}°C is equal to {converted_temp}°F")
    else:
        print("Invalid input for Celsius to Fahrenheit conversion.")
elif user_unit == 'F':
    converted_temp = fahrenheit_to_celsius(user_temp)
    if converted_temp is not None:
        print(f"{user_temp}°F is equal to {converted_temp}°C")
    else:
        print("Invalid input for Fahrenheit to Celsius conversion.")
```

```python
import string

# Implement count_words function
def count_words(text):
    """Counts the number of words in a sentence.

    Args:
        text: The input string.

    Returns:
        The number of words in the string.
    """
    if not isinstance(text, str) or not text.strip():
        return 0 # Handle non-string, empty, or whitespace-only input

    # Remove punctuation and split into words
    text_no_punct = text.translate(str.maketrans('', '', string.punctuation))
    words = text_no_punct.split()

    return len(words)

# Generate and test with test cases
test_cases = [
    ("Hello world", 2),  # Normal text
    ("  Leading and trailing spaces  ", 4), # Leading/trailing spaces - Note: split() handles this
    ("Multiple   spaces   here", 3), # Multiple spaces - Note: split() handles this
    ("Sentence with punctuation!", 3), # Punctuation removed
    ("Another sentence with punctuation, like this.", 5), # More punctuation removed
    ("", 0),  # Empty string
    (" ", 0), # String with only spaces
    ("Word", 1), # Single word
    ("  Word  ", 1), # Single word with spaces
    ("!", 0), # Only punctuation
    ("Hello,world", 1) # No space after comma, treated as one word
]

print("Testing with test cases:")
for text, expected in test_cases:
    result = count_words(text)
    print(f"Input: '{text}', Result: {result}, Expected: {expected}")

# Get user input and test
user_input = input("\nEnter a sentence: ")
user_result = count_words(user_input)

print(f"\nUser Input: '{user_input}'")
print(f"Input: '{user_input}', Result: {user_result}")
```

```python
# Define the BankAccount class with methods and error handling

class BankAccount:
    """A simple bank account class with error handling for invalid operations."""

    def __init__(self, initial_balance=0):
        # Condition: Initial balance must be non-negative
        if not isinstance(initial_balance, (int, float)) or initial_balance < 0:
            raise ValueError("Initial balance must be a non-negative number.")
        self._balance = initial_balance

    def deposit(self, amount):
        """Deposits a positive amount into the account. Raises ValueError for non-positive amounts."""
        # Condition: Negative deposits/withdrawals should raise an error.
        if not isinstance(amount, (int, float)) or amount <= 0:
            raise ValueError("Deposit amount must be a positive number.")
        self._balance += amount
        print(f"Deposited: ${amount:.2f}. New balance: ${self._balance:.2f}")

    def withdraw(self, amount):
        """Withdraws a positive amount from the account. Raises ValueError for non-positive amounts or insufficient funds."""
        # Condition: Negative deposits/withdrawals should raise an error.
        if not isinstance(amount, (int, float)) or amount <= 0:
            raise ValueError("Withdrawal amount must be a positive number.")
        # Condition: Cannot withdraw more than balance.
        if amount > self._balance:
            raise ValueError("Insufficient funds.")
        self._balance -= amount
        print(f"Withdrew: ${amount:.2f}. New balance: ${self._balance:.2f}")

    def check_balance(self):
        """Returns the current account balance."""
        return self._balance

# Generate and Run test cases
print("--- Running Test Cases ---")
# Create a BankAccount instance for testing
account_test = BankAccount(initial_balance=100)

# Test Case 1: Check initial balance
print("\nTest Case 1: Check initial balance")
expected_balance = 100
actual_balance = account_test.check_balance()
print(f"Initial Balance: {actual_balance}, Expected: {expected_balance}")
assert actual_balance == expected_balance

# Test Case 2: Valid Deposit
print("\nTest Case 2: Valid Deposit")
deposit_amount = 50
account_test.deposit(deposit_amount)
expected_balance_after_deposit = 150
actual_balance_after_deposit = account_test.check_balance()
print(f"Balance after deposit: {actual_balance_after_deposit}, Expected: {expected_balance_after_deposit}")
assert actual_balance_after_deposit == expected_balance_after_deposit

# Test Case 3: Valid Withdrawal
print("\nTest Case 3: Valid Withdrawal")
withdraw_amount = 30
account_test.withdraw(withdraw_amount)
expected_balance_after_withdrawal = 120
actual_balance_after_withdrawal = account_test.check_balance()
print(f"Balance after withdrawal: {actual_balance_after_withdrawal}, Expected: {expected_balance_after_withdrawal}")
assert actual_balance_after_withdrawal == expected_balance_after_withdrawal

# Test Case 4: Attempt to deposit a negative amount (should raise ValueError)
print("\nTest Case 4: Attempt to deposit a negative amount")
negative_deposit_amount = -20
try:
    account_test.deposit(negative_deposit_amount)
except ValueError as e:
    print(f"Caught expected error: {e}")
    # Assert that the balance did not change
    assert account_test.check_balance() == expected_balance_after_withdrawal

# Test Case 5: Attempt to withdraw a negative amount (should raise ValueError)
print("\nTest Case 5: Attempt to withdraw a negative amount")
negative_withdraw_amount = -20
try:
    account_test.withdraw(negative_withdraw_amount)
except ValueError as e:
    print(f"Caught expected error: {e}")
    # Assert that the balance did not change
    assert account_test.check_balance() == expected_balance_after_withdrawal

# Test Case 6: Attempt to withdraw more than the current balance (should raise ValueError)
print("\nTest Case 6: Attempt to withdraw more than balance")
insufficient_withdraw_amount = 200
try:
    account_test.withdraw(insufficient_withdraw_amount)
except ValueError as e:
    print(f"Caught expected error: {e}")
    # Assert that the balance did not change
    assert account_test.check_balance() == expected_balance_after_withdrawal

# Test Case 7: Check balance after failed withdrawals
print("\nTest Case 7: Check balance after failed withdrawals")
final_balance = account_test.check_balance()
print(f"Final Balance: {final_balance}, Expected: {expected_balance_after_withdrawal}")
assert final_balance == expected_balance_after_withdrawal

print("\n--- Test Cases Finished ---")

# Get user input and perform operations
print("\n--- User Interaction ---")
user_account = BankAccount(initial_balance=500) # Create a new account for user interaction

while True:
    print("\nChoose an action:")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Check Balance")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == '1':
        try:
            amount = float(input("Enter amount to deposit: "))
            user_account.deposit(amount)
        except ValueError as e:
            print(f"Error: {e}")
    elif choice == '2':
        try:
            amount = float(input("Enter amount to withdraw: "))
            user_account.withdraw(amount)
        except ValueError as e:
            print(f"Error: {e}")
    elif choice == '3':
        print(f"Current balance: ${user_account.check_balance():.2f}")
    elif choice == '4':
        print("Exiting.")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 4.")
```

```python
# Define the function
def is_number_palindrome(num):
    """Checks if an integer reads the same backward as forward.

    Args:
        num: An integer.

    Returns:
        True if the number is a palindrome, False otherwise.
    """
    # Negative numbers are typically not considered palindromes
    if num < 0:
        return False

    # Convert the number to a string
    num_str = str(num)

    # Check if the string reads the same forwards and backwards
    return num_str == num_str[::-1]

# Generate test cases
test_cases = [
    (121, True),    # Positive palindrome
    (-121, False),   # Negative number (not a palindrome in typical definition)
    (10, False),     # Non-palindrome
    (0, True),      # Single digit (considered palindrome)
    (12321, True),  # Longer palindrome
    (12345, False), # Longer non-palindrome
    (1, True),      # Single digit
    (11, True),     # Two-digit palindrome
    (12, False)     # Two-digit non-palindrome
]

# Test the function with test cases
print("Testing with test cases:")
for num, expected in test_cases:
    result = is_number_palindrome(num)
    print(f"Input: {num}, Result: {result}, Expected: {expected}")

# Get user input and test
while True:
    try:
        user_input_str = input("\nEnter an integer to check if it's a palindrome: ")
        user_input = int(user_input_str)
        break # Exit the loop if input is a valid integer
    except ValueError:
        print("Invalid input. Please enter an integer.")


print(f"\nUser Input: {user_input_str}")
user_result = is_number_palindrome(user_input)
print(f"Input: {user_input}, Result: {user_result}")
```
